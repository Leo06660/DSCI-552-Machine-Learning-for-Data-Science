## Part (a) [3.5 points]: 
Find the fattest margin line that separates the points in linsep.txt. Please solve the problem using a Quadratic Programming solver. Report the equation of the line as well as the support vectors.

```python
X = linesp_data[:,:2]
y = linesp_data[:,2]
X[:2]
```

```python
n_samples, n_features = X.shape
```

```python
# P = X^T X
K = np.zeros((n_samples, n_samples))
for i in range(n_samples):
    for j in range(n_samples):
        K[i,j] = np.dot(X[i], X[j])
print(K)
P = cvxopt.matrix(np.outer(y, y) * K)
list(P)
```

```python
# q = -1 (1xN)
q = cvxopt.matrix(np.ones(n_samples) * -1)
q
```

```python
# A = y^T 
A = cvxopt.matrix(y, (1, n_samples))
A
```

```python
# b = 0 
b = cvxopt.matrix(0.0)
b
```

```python
# -1 (NxN)
G = cvxopt.matrix(np.diag(np.ones(n_samples) * -1))
G
```

```python
# 0 (1xN)
h = cvxopt.matrix(np.zeros(n_samples))
h
```

```python
solution = cvxopt.solvers.qp(P, q, G, h, A, b)
```

```python
# Lagrange multipliers
a = np.ravel(solution['x'])
len(a)
```

```python
# Lagrange have non zero lagrange multipliers
sv = a > 0
sv
ind = np.arange(len(a))[sv]
ind
# a = a[sv]
# a
# sv = X[sv]
# len(sv)
# sv_y = y[sv]
```

```python
from cvxopt import matrix, solvers
```

```python
m, n = X.shape
# K = linear_kernel(X, X, degree = 2, intercept = 1)
K = np.matmul(X, X.T)
P = matrix(np.matmul(y,y.T) * K)
q = matrix(np.ones((m, 1)) * -1)
A = matrix(y.reshape(1, -1))
b = matrix(np.zeros(1))          
G = matrix(np.eye(m) * -1)
h = matrix(np.zeros(m))
# list(P)
```

```python
K
```

```python
solution = solvers.qp(P, q, G, h, A, b)
alphas = np.array(solution['x'])
# find the support vectors
ind = (alphas > 1e-4).flatten()
sv = X[ind]
sv_y = y[ind]
alphas = alphas[ind]
```

```python
# print the support vector
print('---alphas---')
print(alphas)
print('-----sv-----')
print(sv)
print('----sv_y----')
print(sv_y)
```

```python
# Calculate w (exclude alpha = 0)
w = np.zeros(n_features)
for n in range(len(alphas)):
    w += alphas[n]*sv_y[n]*sv[n]
w
```

```python
#Converting into cvxopt format
P = cvxopt_matrix(H)
q = cvxopt_matrix(-np.ones((m, 1)))
G = cvxopt_matrix(-np.eye(m))
h = cvxopt_matrix(np.zeros(m))
A = cvxopt_matrix(y.reshape(1, -1))
b = cvxopt_matrix(np.zeros(1))
```

# Part (b) [3.5 points]: 
Using a kernel function of your choice along with the same Quadratic Programming solver, find the equation of a curve that separates the points in nonlinsep.txt. Report the kernel function you use as well as the support vectors.

```python
nonlinesp_data = []
with open('nonlinsep.txt') as f:
    file = csv.reader(f)
    for line in file:
        nonlinesp_data.append(line)
nonlinesp_data = np.array(nonlinesp_data).astype('float64')
nonlinesp_data[:6]
```

```python
X = nonlinesp_data[:,:2]
y = nonlinesp_data[:,2]
X[:2]
n_samples, n_features = X.shape
```

```python
def RBF_kernel_function(x1, x2, gamma=0.01):
    return np.exp(-gamma * np.linalg.norm(x1 - x2) ** 2)
```

```python
# new Q from kernel function(K)
K = np.zeros((n_samples, n_samples))
for i in range(n_samples):
    for j in range(n_samples):
        K[i,j] = RBF_kernel_function(X[i], X[j])
print(K)
# P = cvxopt.matrix(np.outer(y, y) * K)
# list(P)
```

```python
m, n = X.shape
P = matrix(np.matmul(y,y.T) * K)
q = matrix(np.ones((m, 1)) * -1)
A = matrix(y.reshape(1, -1))
b = matrix(np.zeros(1))          
G = matrix(np.eye(m) * -1)
h = matrix(np.zeros(m))
```

```python
solution = solvers.qp(P, q, G, h, A, b)
alphas = np.array(solution['x'])
# find the support vectors
ind = (alphas > 1e-4).flatten()
sv = X[ind]
sv_y = y[ind]
alphas = alphas[ind]
```

```python
# print the support vector
print('---alphas---')
print(alphas)
print('-----sv-----')
print(sv)
print('----sv_y----')
print(sv_y)
```
